## AVERAGE. Per region, compute percentage of population with credit card AND debit card segmented by gender.

Hint:
Columns will be region, % of male with BOTH credit card AND debit card, % of female with BOTH credit card AND debit card

```python
#group first per region
#per region get the population with credit and debit
#

# Create new column
data['with debit card AND credit card'] = data.apply(
    lambda x: 1 if x['fin2'] == 1 and x['fin7'] == 1 else None, 
    axis=1
)
```

```python
# Show data
data[['fin2','fin7', 'with debit card AND credit card']]
```

```python
# Apply groupings
grouped_data = data.groupby(['regionwb','female'])['wpid_random', 'with debit card AND credit card'].count()
grouped_data = grouped_data.reset_index()
```

```python
# Show data
grouped_data
```

```python
# Compute % population with credit card AND debit card
grouped_data['% of population with debit card AND credit card']=grouped_data['with debit card AND credit card']*100.0/grouped_data['wpid_random']
```

```python
# Show data
grouped_data
```

```python
# Filter data
grouped_data = grouped_data[
    [
        'regionwb',
        'female',
        '% of population with debit card AND credit card'
    ]
]
```

```python
# Show data
grouped_data
```

```python
# Apply pivot
pivoted_data = grouped_data.pivot(
    index ='regionwb', 
    columns ='female', 
    values =['% of population with debit card AND credit card']
    ).reset_index()
```

```python
# Show data
pivoted_data
```

## DIFFICULT. Compute the following metric per region:

card ownership access gender ratio = % of females with BOTH credit card AND debit card/% of males with BOTH credit card AND debit card

No apparent gender discrepancy if metric approaches 1. Else, there is a discrepancy.

```python
# Access each column of the multiindex dataframe and save it to a variable
females = pivoted_data['% of population with debit card AND credit card', 2]
males = pivoted_data['% of population with debit card AND credit card', 1]
regions = pivoted_data['regionwb', '']
```

```python
# Create dictionary of Series data
series_dict = {
    'regions': regions,
    '% of female population with debit card AND credit card': females,
    '% of male population with debit card AND credit card': males
}
```

```python
# Create a new dataframe
discrepancy_data = pd.DataFrame(series_dict)
```

```python
# Show data
discrepancy_data
```

```python
# Compute card ownership access gender ratio
discrepancy_data['card ownership access gender ratio'] = discrepancy_data['% of female population with debit card AND credit card']/discrepancy_data['% of male population with debit card AND credit card']
```

```python
# Sort (from highest to lowest ratio). Note that ratio approaching 1 is better!
discrepancy_data = discrepancy_data.sort_values(by=['card ownership access gender ratio'], ascending=[False])
```
